# Modelo de clasificación supervisada para categorizar la Calidad de vino verde rojo y blanco

## Objetivo del Estudio

El presente estudio se centra en la base de datos `calidad-del-vino-blanco.csv` y `calidad-del-vino-rojo.csv`, que incluye información detallada de algunas caracteristicas fisicoquímicas del vino verde blanco y rojo y la calidad asosiada a cada muestra que se evaluo, como:

- **Acidez fija** (`fixed acidity`)
- **Acidez volátil** (`volatile acidity`)
- **Ácido cítrico** (`citric acid`)
- **Azúcar residual** (`residual sugar`)
- **Cloruros** (`chlorides`)
- **Dióxido de azufre libre** (`free sulfur dioxide`)
- **Dióxido de azufre total** (`total sulfur dioxide`)
- **Densidad** (`density`)
- **PH** (`pH`)
- **Sulfatos** (`sulphates`)
- **Alcohol** (`alcohol`)
- **Calidad (puntuación entre 0 y 10)** (`quality`)

# Analisis Exploratorios

```python
# Definir una paleta de colores consistente para cada categoría de calidad
unique_qualities = sorted(df['quality'].unique())  # ordenar las categorias en orden ascedente
palette = sns.color_palette('tab10', len(unique_qualities)) # generar paleta de colores por grupo

def plot_group_stats(df, category_column=None, columns_to_exclude=None):
    """
    Genera gráficos descriptivos en una única fila por característica,
    colocando el boxplot al inicio seguido de los histogramas por categoría.

    Args:
    df (DataFrame): DataFrame que contiene los datos a graficar.
    category_column (str, opcional): Nombre de la columna categórica para usar en la diferenciación por colores.
    columns_to_exclude (list, opcional): Lista de nombres de columnas que deben excluirse de los gráficos.
    """

    # Definir las columnas a excluir si no se proporciona ninguna lista
    if columns_to_exclude is None:
        columns_to_exclude = ['quality']

    # Si se proporciona una columna categórica, agregarla a la lista de exclusión
    if category_column and category_column not in columns_to_exclude:
        columns_to_exclude.append(category_column)

    # Seleccionar las columnas que se van a graficar
    columns_to_plot = [col for col in df.columns if col not in columns_to_exclude]

    # Configuración del grid con una única fila por característica
    num_features = len(columns_to_plot)
    fig, axes = plt.subplots(num_features, len(unique_qualities) + 1, figsize=(25, 4 * num_features))

    # Bucle para recorrer las características
    for i, column in enumerate(columns_to_plot):

        # Subplot para el boxplot en la primera columna
        ax = axes[i, 0]  # Boxplot en la primera posición de la fila

        # Utilizar el parámetro hue si se especifica una categoría
        if category_column:
            sns.boxplot(x='quality', y=column, hue=category_column, data=df, ax=ax)
        else:
            sns.boxplot(x='quality', y=column, data=df, palette=palette, ax=ax)

        ax.set_title(f'Boxplot de {column}', fontsize=12)
        ax.set_ylabel(column, fontsize=12)
        ax.grid()

        # Subplots para histogramas en las siguientes columnas
        for j, quality_value in enumerate(unique_qualities):
            ax = axes[i, j + 1]  # Ajuste para histogramas en las columnas restantes

            # Filtrar la variable objetivo por categoria y graficar
            if category_column:
                sns.histplot(
                    df[df['quality'] == quality_value],
                    x=column,
                    hue=category_column,
                    kde=True,
                    ax=ax,
                    bins=15
                )
            else:
                sns.histplot(
                    df[df['quality'] == quality_value][column],
                    kde=True,
                    ax=ax,
                    color=palette[j],
                    bins=15
                )

            ax.grid()
            if i == 0:  # Títulos solo en la primera fila
                ax.set_title(f'Quality {quality_value}', fontsize=12)

    plt.tight_layout()
    plt.show()

# Instanciar la función para generar gráficos
plot_group_stats(df, category_column='tipo vino', columns_to_exclude=['quality', 'tipo vino'])
```

# División en la data en entrenamiento y testeo

```python
# Definimos las variables categóricas y numéricas
categorical_features = x.select_dtypes(include=['object']).columns
numerical_features = x.select_dtypes(include=['float64']).columns

# Creamos pipelines de preprocesamiento
numerical_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore',sparse_output=False))
])

# Usamos ColumnTransformer para combinar ambas transformaciones
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
# Definimos el pipeline completo
model_pipeline = Pipeline(steps=[('preprocessor', preprocessor)])
```

```python
model_pipeline.fit_transform(x)
```

```python
model_pipeline.fit_transform(x).shape
```

```python
# Dividimos los datos en entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, shuffle=True, stratify=y)
```

# Aplicación de modelos de clasificación supervisada
- Gradient Boosting: Se seleciono este modelo ya que una de sus ventajas es que trabaja bien con datos desaquilibrados y según la ifnormación analisada y entregada es una de las caracterizticas de este data set
- Vecinos más Cercanos (K-Nearest Neighbors, K-NN): Teniendo en cuenta que estamos hablando de una variable como la calidad de un producto esta en general se define bajo ciertos rangos que tiene que cumplir el producto analizado para cumplir con los estandares, por lo cual podria tener execentes resultados teniendo en cuenta que las variables de entrada evaluadas tendrian en teoria que cumplir con un rango para tener alguno de los puntajes de calidad.
- Random Forest: Al tener cierta similitud con el modelo grandient boosting, seria interesante comparar los resultados de ambos modelos y analizar sus cambios acorde a sus diferencias teoricas partiendo de esta similitud.


```python
def grafico_evolucion_entrenamiento(y_test, y_pred, model, X_train, y_train, X_val, y_val):
    # Crear figura con dos subgráficos en una sola fila
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # Subgráfico 1: Matriz de Confusión
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False, ax=axes[0])
    axes[0].set_xlabel('Predicción')
    axes[0].set_ylabel('Valor Real')
    axes[0].set_title('Matriz de Confusión')
    axes[0].grid(True)

    # Variables para almacenar la evolución del error
    train_errors = []
    test_errors = []

    # Registro del error en cada etapa del entrenamiento incremental
    for y_pred_train in model.staged_predict(X_train):
        train_errors.append(1 - accuracy_score(y_train, y_pred_train))

    for y_pred_val in model.staged_predict(X_val):
        test_errors.append(1 - accuracy_score(y_val, y_pred_val))

    # Subgráfico 2: Evolución del Error
    axes[1].plot(np.arange(1, len(train_errors) + 1), train_errors, label='Error de Entrenamiento', linestyle='--')
    axes[1].plot(np.arange(1, len(test_errors) + 1), test_errors, label='Error de Prueba')
    axes[1].set_xlabel('Número de Estimadores')
    axes[1].set_ylabel('Error')
    axes[1].set_title('Evolución del Error durante el Entrenamiento')
    axes[1].grid(True)
    axes[1].legend()

    # Mostrar la figura
    plt.tight_layout()
    plt.show()
```

# Gradient Boosting

## Concepto Matemático de Gradient Boosting

El **Gradient Boosting** es una técnica de aprendizaje supervisado que combina varios modelos débiles (como árboles de decisión) en una secuencia para crear un modelo más robusto. Su idea principal es optimizar una función de pérdida mediante el método del **gradiente descendente**.

### 1. Modelo Secuencial

El modelo general tras \(m\) iteraciones se representa como:

$$
F_{m}(x) = F_{m-1}(x) + \eta \cdot h_m(x)
$$

Donde:
- \(F_m(x)\) es el modelo acumulado tras \(m\) iteraciones.
- \(h_m(x)\) es el modelo en la \(m\)-ésima iteración que intenta minimizar los errores.
- \(\eta \in (0, 1]\) es la **tasa de aprendizaje**, un hiperparámetro que controla cuánto contribuye el nuevo modelo al modelo acumulado.

### 2. Minimización del Error

Cada nuevo modelo \(h_m(x)\) se ajusta para minimizar el residuo (error) que dejó el modelo anterior. Los residuos se calculan como:

$$
r_i^{(m)} = y_i - F_{m-1}(x_i)
$$

Aquí:
- \(r_i^{(m)}\) es el residuo del \(i\)-ésimo dato en la iteración \(m\).
- \(y_i\) es el valor real del \(i\)-ésimo dato.
- \(F_{m-1}(x_i)\) es la predicción del modelo anterior para el \(i\)-ésimo dato.

### 3. Optimización con Gradiente Descendente

Gradient Boosting busca minimizar una **función de pérdida** \(L(y, F(x))\), típicamente el **error cuadrático medio** en problemas de regresión:

$$
L(y, F(x)) = \frac{1}{n} \sum_{i=1}^{n} \left( y_i - F(x_i) \right)^2
$$

El gradiente de esta función con respecto a las predicciones \(F(x_i)\) es:

$$
\frac{\partial L}{\partial F(x_i)} = -2 \cdot \left( y_i - F(x_i) \right)
$$

Este gradiente indica la dirección en la que debe ajustarse el nuevo modelo para corregir los errores del modelo anterior.

---

En resumen, el algoritmo **Gradient Boosting** entrena los modelos de manera secuencial, utilizando los residuos (errores) del modelo anterior como la guía para el siguiente. Cada nuevo modelo intenta mejorar sobre los errores previos, y la optimización se realiza mediante **gradiente descendente** para minimizar una función de pérdida.

# **Primer Entrenamiento**

```python
# Creamos un nuevo pipeline que incluye el preprocesamiento y el modelo
model_pipeline_GB = Pipeline(steps=[('preprocessor', preprocessor),
                                 ('classifier', GradientBoostingClassifier(n_estimators=1000, learning_rate=1.0, max_depth=5, random_state=0))])

# Entrenamos el pipeline completo en los datos de entrenamiento
model_pipeline_GB.fit(x_train, y_train)

# Hacemos predicciones en el conjunto de prueba
y_pred = model_pipeline_GB.predict(x_test)
gb_model = model_pipeline_GB.named_steps['classifier']

# Evaluamos el rendimiento del modelo
print(classification_report(y_test, y_pred, zero_division='warn'))

# Preprocesamiento de los datos de entrada
X_train_processed = model_pipeline_GB.named_steps['preprocessor'].transform(x_train)
X_test_processed = model_pipeline_GB.named_steps['preprocessor'].transform(x_test)

# Extraer el modelo GradientBoostingClassifier desde el Pipeline
gb_model = model_pipeline_GB.named_steps['classifier']

# Llamar a la función con los datos preprocesados
grafico_evolucion_entrenamiento(y_test, y_pred, gb_model, X_train_processed, y_train, X_test_processed, y_test)
```

# **Analisis y manejo de los resultados**
Para este primer entrenamiento se determina que un `n_estimators=1000` es muy alto dado que el gradiente dismuye hacia el `n_estimators=200` dejando de aprender, y con el objetivo de mejorar el rendimiento del modelo se hace una seguna parmetrizacion con estos parametros:
`GradientBoostingClassifier(n_estimators=500, learning_rate=0.05, max_depth=3, random_state=42)`

# **Comparación de Parámetros Gradient Boosting**

| **Parámetro**            | **Configuración Propuesta**                                         | **Configuración Actual**                                        | **Descripción Simple** |
|--------------------------|-------------------------------------------------------------|------------------------------------------------------------|------------------------|
| **`n_estimators`**       | 500                                                         | 1000                                                       | Número de árboles en el modelo. Más árboles pueden mejorar el rendimiento, pero aumentan el tiempo de entrenamiento y el riesgo de sobreajuste. |
| **`learning_rate`**      | 0.05                                                        | 1.0                                                        | Controla cuánto cambia el modelo en cada iteración. Un valor más bajo hace que el modelo aprenda lentamente, mientras que un valor más alto acelera el aprendizaje, pero puede llevar a sobreajuste. |
| **`max_depth`**          | 3                                                           | 5                                                          | Define la profundidad máxima de cada árbol. Árboles más profundos capturan patrones más complejos, pero también aumentan el riesgo de sobreajuste. |
| **`random_state`**       | 42                                                          | 0                                                          | Fija la semilla para la generación aleatoria, permitiendo reproducir los resultados. |


> **Ventaja Principal :**   Menor riesgo de sobreajuste debido al aprendizaje lento. Puede capturar patrones más complejos debido a más árboles y mayor profundidad.

> **Desventaja Principal :**  Podría requerir más iteraciones para converger. Mayor riesgo de sobreajuste debido al rápido aprendizaje y más parámetros ajustables.

---

## **Conclusiones de los Parámetros**

1. **`n_estimators` (Número de Árboles):**
   - **Más árboles (Configuración 2)** pueden capturar más patrones en los datos, pero aumentan el tiempo de entrenamiento.
   - **Menos árboles (Configuración 1)** pueden ser más rápidos, pero con menos capacidad para capturar patrones complejos.

2. **`learning_rate` (Tasa de Aprendizaje):**
   - Un **valor bajo (Configuración 1)** hace que el modelo aprenda más lentamente, lo que reduce el riesgo de sobreajuste.
   - Un **valor alto (Configuración 2)** acelera el aprendizaje, pero puede hacer que el modelo sobreajuste los datos si no se controla adecuadamente.

3. **`max_depth` (Profundidad del Árbol):**
   - **Árboles más profundos (Configuración 2)** permiten capturar relaciones más complejas en los datos.
   - **Árboles menos profundos (Configuración 1)** simplifican el modelo, lo que puede ser útil para evitar sobreajuste.

4. **`random_state` (Semilla Aleatoria):**
   - Este parámetro solo se utiliza para asegurarse de que los resultados sean reproducibles al fijar la semilla aleatoria para la generación de datos.

# **Conceptos de SMOTE y SMOTEENN para Modelos Ajustados**

## **1. ¿Qué es SMOTE?**
**SMOTE (Synthetic Minority Oversampling Technique)** es una técnica de **sobremuestreo** que se utiliza para abordar el problema del **desequilibrio de clases** en conjuntos de datos.
- **Objetivo:** Crear ejemplos sintéticos de las clases minoritarias para equilibrar las clases.
- **Cómo funciona:** Genera nuevos ejemplos interpolando entre puntos existentes de la clase minoritaria y sus vecinos más cercanos.
  
**Ventaja:** Mejora el rendimiento del modelo al darle más ejemplos de las clases minoritarias.

---

## **2. ¿Qué es SMOTEENN?**
**SMOTEENN** es una combinación de **SMOTE** y **Edited Nearest Neighbors (ENN)**. Esta técnica realiza una **combinación de sobremuestreo y submuestreo**:
1. **Primero, SMOTE** genera ejemplos sintéticos para las clases minoritarias.
2. **Después, ENN** elimina muestras (tanto de la clase mayoritaria como de la minoritaria) que están mal clasificadas por sus vecinos más cercanos.

**Ventaja:**
- Mejora la **calidad de los datos** al eliminar ruido y ejemplos mal etiquetados.
- Equilibra el conjunto de datos de manera más efectiva que solo SMOTE.

---

## **3. ¿Por qué Usar SMOTE y SMOTEENN?**
- En datasets con **desequilibrio de clases**, los modelos tienden a **favorecer la clase mayoritaria**, ignorando las minoritarias.
- SMOTE **aumenta las instancias** de las clases minoritarias para que el modelo tenga más ejemplos de estas.
- SMOTEENN no solo aumenta los ejemplos minoritarios, sino que también **reduce el ruido**, mejorando así la calidad del aprendizaje.

---

## **4. Cuándo Utilizar Cada Técnica**
- **SMOTE:** Si el conjunto de datos tiene pocas clases minoritarias y el enfoque es principalmente en **agregar más ejemplos** de estas clases.
- **SMOTEENN:** Si el dataset tiene **ruido significativo** o muchos ejemplos mal etiquetados. Esta técnica proporciona un **balance más limpio** y efectivo.

---

## **Conclusión**
Ambas técnicas son útiles para mejorar el rendimiento de los modelos en datasets desbalanceados. **SMOTE** es más simple y directo, mientras que **SMOTEENN** proporciona un enfoque más sofisticado al combinar sobremuestreo y limpieza de datos. Estas técnicas garantizan que el modelo no solo aprenda patrones de la clase mayoritaria, sino también de las clases minoritarias, mejorando la **generalización** y la **precisión del modelo**.


```python
from imblearn.combine import SMOTEENN
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
# 1. Preprocesar los datos de entrenamiento
X_train_processed = preprocessor.fit_transform(x_train)  # Asegurarse de que X_train esté procesado
X_test_processed = preprocessor.transform(x_test)

# 2. Crear una instancia de SMOTE con k_neighbors ajustado
smote = SMOTE(random_state=42, k_neighbors=2)

# 3. Aplicar SMOTEENN con la instancia de SMOTE ajustada
smote_enn = SMOTEENN(smote=smote, random_state=42)
X_train_balanced, y_train_balanced = smote_enn.fit_resample(X_train_processed, y_train)

# 4. Verificar la nueva distribución de las clases
print("Distribución después de SMOTEENN:")
print(pd.Series(y_train_balanced).value_counts())

# 5. Dividir el conjunto de entrenamiento para validación
X_train_final, X_val, y_train_final, y_val = train_test_split(
    X_train_balanced, y_train_balanced, test_size=0.2, random_state=42
)

# 6. Entrenar el modelo con los datos balanceados
model = GradientBoostingClassifier(random_state=42, n_estimators=100)
model.fit(X_train_final, y_train_final)

# 7. Hacer predicciones sobre el conjunto de prueba
y_pred = model.predict(X_test_processed)

# 8. Evaluación del modelo
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall (Sensibilidad): {recall:.2f}")
print(f"F1-Score: {f1:.2f}")

# Reporte de clasificación detallado
print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred))

# 9. Graficar los resultados con la función personalizada
grafico_evolucion_entrenamiento(y_test, y_pred, model, X_train_final, y_train_final, X_val, y_val)
```

```python
# Importar las bibliotecas necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix

# Generar datos simulados
np.random.seed(42)
data = {
    'acidez_fija': np.random.uniform(4, 15, 100),
    'acidez_volatil': np.random.uniform(0, 1.5, 100),
    'acido_citrico': np.random.uniform(0, 1, 100),
    'azucar_residual': np.random.uniform(0, 20, 100),
    'cloruros': np.random.uniform(0, 0.1, 100),
    'sulfatos': np.random.uniform(0.4, 1.5, 100),
    'alcohol': np.random.uniform(8, 14, 100),
    'calidad': np.random.randint(3, 9, 100)  # Etiquetas entre 3 y 8
}

# Crear el DataFrame
df = pd.DataFrame(data)

# Separar las características y etiquetas
X = df.drop('calidad', axis=1)
y = df['calidad']

# División del dataset en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Definir el modelo y los parámetros a optimizar
model = GradientBoostingClassifier()
param_grid = {'n_estimators': [50, 100], 'learning_rate': [0.01, 0.1], 'max_depth': [3, 5]}

# Realizar la búsqueda de hiperparámetros con GridSearchCV
grid_search = GridSearchCV(model, param_grid, cv=2, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)

# Evaluar el mejor modelo
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# Reporte de clasificación
print("Mejores Hiperparámetros:", grid_search.best_params_)
print("Reporte de Clasificación:")
print(classification_report(y_test, y_pred))

# Graficar la matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)
plt.matshow(conf_matrix, cmap='Blues')
plt.colorbar()
plt.xlabel('Predicción')
plt.ylabel('Actual')
plt.title('Matriz de Confusión')
plt.show()

print(confusion_matrix(y_test, y_pred))
```
